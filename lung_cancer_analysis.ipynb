```python
# Assuming 'df' is your DataFrame
df = pd.read_csv("/content/sample_data/cancer patient data sets.csv")
```

```python

df.plot(kind='scatter', x='Level', y='Passive Smoker', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Assuming df is your DataFrame
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 14))
fig.suptitle('Histograms for Each Column', fontsize=16)

# Flatten the axes to iterate through them
axes = axes.flatten()

# Plot histograms for each column up to the number of available subplots
for i, col in enumerate(df.columns):
    if i < len(axes):
        df[col].hist(ax=axes[i], bins=20)
        axes[i].set_title(col)

# Adjust layout and spacing
plt.subplots_adjust(hspace=0.5, bottom=0.1)

plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, x='Age', hue='Level', kde=True, palette='Set2', element='step', common_norm=False)
plt.title('Distribution of Age by Level')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Level')

plt.show()
```
